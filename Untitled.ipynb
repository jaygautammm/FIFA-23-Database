```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#how overall affects different aspects of game

sns.set_palette("Set1")
fig = plt.gcf()
plt.ylabel("Performance")

sns.lineplot(x = "Overall", y = "Finishing", data = data, label = "Finishing",legend = 'brief')
sns.lineplot(x = "Overall", y = "Passing", data = data, label = "Passing",legend = 'brief')
sns.lineplot(x = "Overall", y = "Defending", data = data, label = "Defending",legend = 'brief')

plt.show()
```

```python
#As per the above analysis players with higher overall have higher variation in their attributes
```

```python
sns.set_palette("Set2")
fig = plt.gcf()
plt.ylabel("Physicality")

sns.lineplot(x = "Age", y = "Sprint", data = data, label = "Sprint", legend = 'brief')
sns.lineplot(x = "Age", y = "Acceleration", data = data, label = "Acceleration", legend = 'brief')
sns.lineplot(x = "Age", y = "Agility", data = data, label = "Agility", legend = 'brief')
sns.lineplot(x = "Age", y = "Aggression", data = data, label = "Aggression", legend = 'brief')
sns.lineplot(x = "Age", y = "Stamina", data = data, label = "Stamina", legend = 'brief')
sns.lineplot(x = "Age", y = "Jumping", data = data, label = "Jumping", legend = 'brief')

plt.show()
```

```python
#As per the above analysis players physicality seems to peak at 24-28 years of age
```
